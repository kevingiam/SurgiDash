```python
import pandas as pd
import plotly.figure_factory as ff
```

```python
date_columns = ['p_birth_date', 'p_admission_date', 'p_surgery_date', 'p_separation_date', 'p_death_date']
df_raw = pd.read_csv('/Users/kevingiam/Aspera/Home/pcore_clean_2022_11_03.csv', parse_dates=date_columns)
```

```python
hist_data = []
h_names = df_raw['h_name'].unique()
for h_name in h_names:
    hist_data.append(df_raw.loc[df_raw['h_name'] == h_name, 'p_asa_ps'].values)
```

```python
fig = ff.create_distplot(hist_data, h_names)

fig.update_layout(
    width=800, 
    height=500,
    legend=dict(
        y=-0.15,
        yanchor='top',
        orientation="h",
    ),
    margin=dict(l=20, r=20, t=20, b=20)
)

fig.show()
```

Example: ASA-PS value should be between 1 and 5

The plot above visualise the distribution of ASA-PS values accross 3 hospitals. We can see that the distribution of ASA-PS values in Peter Maccallum and Flinders Medical Centre are similar. However, those distributions are very different with Alfred Health. This indicates that there are some error in the data and the semi automated algorithm should raise a red flag to the data engineer.
